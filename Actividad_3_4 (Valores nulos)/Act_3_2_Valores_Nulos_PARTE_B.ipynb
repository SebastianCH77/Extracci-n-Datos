```python
#Leer el archivo
df = pd.read_csv('Base_Socio_Formador_Con_Nulos.csv', encoding="latin")
```

```python
#Mostrar las primeras 15 filas incluyendo todas las columnas
df.head(15)
```

```python
#Verificar el tipo de datos de cada columna para su posterior análisis
df.info()
```

```python
#Se identifican valores nulos por cada columna
valores_nulos=df.isnull().sum()
valores_nulos
```

```python
#Realizamos una copia del dataframe para guardarlo posteriormente y mantener el original aparte
df2=df.copy()
```

**FECHA**

```python
#Averiguando cuál es la fecha más reciente de la columna "fecha" utilizando el formato de datetime
df2['fecha'] = pd.to_datetime(df2['fecha'])
df2['fecha'].max()
```

```python
#Aplicando el método de sustitución de valores nulos por un string en concreto, en este caso se tomará la fecha más reciente ya que puede ser la más representativa para un análisis reciente
df2["fecha"] =df2["fecha"].fillna("2019-08-15 00:00:00")
```

**PRECIO**

```python
#Para la columna de 'precio' empezamos por convertir a formato numérico quitando algunos símbolos como '$' y ','
df['precio'] = pd.to_numeric(df['precio'].replace({'\$': '', ',': ''}, regex=True), errors='coerce')
```

```python
 #Dado que los precios tienen una distribución sesgada con minimos y máximos se aplicará el método de sustitución de valores nulos con la mediana, esto permite no verse afectada por valores atípicos
df2["precio"]=df2["precio"].fillna(round(df["precio"].median(),1))
```

**AÑO HECHO**

```python
#Verificando los valores de la columna año_hecho
df2['año_hecho'].value_counts()
```

```python
#Se aplica el método de sustitución de valores nulos por un string en concreto, en este caso por la fecha más repetida que fue 01/01/2005
df2["año_hecho"] =df2["año_hecho"].fillna("01/01/2005")
```

**AÑO RENOVADO**

```python
#Para la columna año_renovado los valores nulos podrían tener significado si la la propiedad no se ha renovado, por lo que sustituirá por un valor numérico en concreto, en este caso "0"
df2["año_renovado"] =df2["año_renovado"].fillna(0.0)
```

**MTS CUADRADOS**

```python
#Se alica el método de sustitución de valores nulos con la media de la columna metros_cuadrados ya que el área puede tener variaciones importantes
df2["metros_cuadrados"]=df2["metros_cuadrados"].fillna(round(df["metros_cuadrados"].mean(),1))
```

**PISOS**

```python
#Comprobando el número de pisos que se repiten más en las propiedades
df2["pisos"].mode().iloc[0]
```

```python
# Debido a que las propiedades suelen tener un número limitado de pisos y 1 es el valor más repetio es 2 se aplica el método de sustitución de valores nulos por un valor numérico en  concreto, en este caso 2.0
df2["pisos"]= df2["pisos"].fillna(2.0)
```

**HABITACIONES**

```python
#Comprobando el número de habitacines que se repiten más en las propiedades
df2["habitaciones"].mode().iloc[0]
```

```python
#De manera similar, el valor más repetido en número de habitaciones es 3, se aplica el método de sustitución de valores nulos por un valor numérico en  concreto, en este caso 3.0
df2["habitaciones"]= df2["habitaciones"].fillna(3.0)
```

**BAÑOS**

```python
#Comprobando el número de baños que se repiten más en las propiedades
df2["baños"].mode().iloc[0]
```

```python
#De manera similar, el valor más repetido en número de baños es 3, se aplica el método de sustitución de valores nulos por un valor numérico en  concreto, en este caso 3.0
df2["baños"]= df2["baños"].fillna(3.0)
```

**CALLE**

```python
#Para la calle, al ser una dirección muy particular se opta por mantener los valores nulos utilizando el método de sustitución de valores nulos por un string llamado "Sin dirección"
df2["calle"] =df2["calle"].fillna("SIN DIRECCION")
```

**ZIP**

```python
#Verificando cuál es el ZIP que más se repite y cuántas veces
valor_mas_repetido = df2["codigo_zip"].mode().iloc[0]
frecuencia = df2["codigo_zip"].value_counts().max()

print(f"ZIP más repetido: {valor_mas_repetido}")
print(f"Se repite {frecuencia} veces")
```

```python
#Empleando el método de sustitución de valores nulos por un valor numérico en  concreto, en este caso asignaremos el ZIP 75217.0 ya que fue el más repetido
df2["codigo_zip"]= df2["codigo_zip"].fillna(75217.0)
```

**ESTADO**

```python
#Se hace un filtro para comprobar a qué estado y ciudad pertenece el ZIP número 75217 para hacer la sustitución
df2[df2["codigo_zip"] == 75217.0]
```

```python
#Se hace la sutitución del valor nulo en estado por TX ya que es el estado donde se encuentra el ZIP 75217
df2["estado"] =df2["estado"].fillna("TX")
```

**CIUDAD**

```python
#Finalmente se hace la sutitución del valor nulo en ciudad por DALLAS ya que es el estado donde se encuentra el ZIP 75217
df2["ciudad"] =df2["ciudad"].fillna("DALLAS")
```

```python
#Corroboramos valores nulos
valores_nulos=df2.isnull().sum()
valores_nulos
```
