```python
#Carga desde un archivo .xlsx sin indice
df= pd.read_excel('Datos de Facturación.xlsx')
#df
```

```python
#Verificamos información del DataFrame
df.info()
```

```python
#Imprimo los primeros 5 registro del dataframe
df.head()
```

```python
#Conversión de tipo de dato de columna de tipo Object a Float
df['FECHA_ENT']= df['FECHA_ENT'].astype(str)
```

```python
#Reemplazo de datos tipo string a datos tipo numérico
df.CVE_CLPV = df.CVE_CLPV.replace({'M':'0'}, regex=False)
```

```python
filtro1 = df[(df['CVE_CLPV'] >= 1000) & (df['CVE_CLPV'] <= 2000)]
filtro1
```

```python
#Todas las claves excepto 5 y 4 de CVE_VEND
#Filtro not
filtro2=df[~(df["CVE_VEND"] == 5) & ~(df["CVE_VEND"] == 4)]
filtro2
```

```python
#De la columna FECHA_ENT filtrar la fecha 2022-02-28 00:00:00
filtro3=df[df["FECHA_ENT"] == "2022-02-28 00:00:00"]
filtro3
```

```python
#Cantidades menores a 5951.7 de CAN_TOT ó STATUS "E"
#Filtro o
filtro4=df[(df["CAN_TOT"] < 5951.7) | (df["STATUS"] == "E")]
filtro4
```

```python
#Sólo las columnas 	CVE_DOC	FECHA_ENT	FECHA_VEN	CAN_TOT
#Filtro por columnas
filtro5=df.iloc[ : , [0,6,7,9]]  #Columnas no consecutivas
filtro5
```

```python
#Sólo las filas de 7001 a 7099
#Filtro por filas
filtro6= df.iloc[7001:7100 , : ]           #Filas consecutivas
filtro6
```

```python
filtro7 = df[(df['CVE_VEND'] == 1) | (df['CVE_VEND'] == 2) ]
filtro7 = filtro7.set_index('CVE_VEND')
filtro7['FECHAELAB']
```
