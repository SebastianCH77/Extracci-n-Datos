```python
#Carga desde un archivo .xlsx sin indice
df= pd.read_excel('Detalle precios y productos fabricados 2022.xlsx')
#df
```

```python
#Verificamos información del DataFrame
df.info()
```

```python
#Imprimo los primeros 5 registro del dataframe
df.head()
```

```python
#Filtro por objeto (De la columna vendedor filtrar las operaciones de la vendedora Leticia Ramirez)
filtro1=df[df["NOMBRE_VENDEDOR"] == "LETICIA RAMIREZ HERNANDEZ"]
filtro1
```

```python
#Filtro por filas consecutivas (Filtrar en el índice del 5 al 11)
filtro2= df.iloc[5:11 , : ]           #Filas consecutivas
filtro2
```

```python
#Filtro por columnas no consecutivas (Filtrando todos los valores del índice con las columnas 1, 4 y 8 )
filtro3=df.iloc[ : , [1,4,8]]  #Columnas no consecutivas
filtro3
```

```python
df1= pd.read_excel('Detalle precios y productos fabricados 2022.xlsx', index_col=0)
df1
```

```python
#Paso 2: Aplicamos Filtro de filas vs columnas (Tomando en cuenta el número de la clave y la cantidad)
filtro4=df1.loc[["F37034","F37041"], ["CANT"]]
filtro4
```

```python
#Filtro por cabecera
filtro5=df.tail()
filtro5
```

```python
#Filtro por comparacion (Con cantidad menor o igual a l0)
filtro6=df[df["CANT"] <= 10]
filtro6
```

```python
#Filtro Y (Filtrando con cantidad y por fecha doc con la fecha "2022-01-05")
filtro7=df[(df["CANT"] < 5) & (df["FECHA_DOC"] == "2022-01-05")]
filtro7
```

```python
#Filtro o (Por cantidad ó por fecha)
filtro8=df[(df["CANT"] < 6) | (df["FECHA_DOC"] == "2022-01-05")]
filtro8
```

```python
#Filtro not (Negando las condicionales del filtro, en este caso esa fecha)
filtro9=df[~(df["FECHA_DOC"] == "2022-12-23")]
filtro9
```

```python
filtro9.to_csv("filtro9.csv")
```
