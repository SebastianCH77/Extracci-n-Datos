```python
#Carga desde un archivo .xlsx sin indice
df= pd.read_excel('Inmuebles.xlsx')
#df
```

```python
#Eliminar primera fila o filas innecesarias
df= pd.read_excel('Inmuebles.xlsx', skiprows=1)
df
```

```python
#Eliminar columnas innecesarias
#axis = 1 para columnas
#axis = 0 para filas
df= df.drop(['Referencia'], axis=1)
df
```

**INCISO B TODOS LOS REGISTROS DE FECHA ALTA IGUAL A 2007-04-29 Y 2004-01-02**

```python
#Aplicando un filtro Ó para ambas fechas con la misma columna
filtrob = df[(df["Fecha_Alta"] == "2007-04-29 00:00:00") | (df["Fecha_Alta"] == "2004-01-02 00:00:00")]
filtrob
```

**INCISO C TODOS LOS REGISTROS DE "Tipo: parking, local, industrial y piso"**

```python
#Aplicando un filtro CON .isin ya que LOS VALORES son de tipo objeto
filtroc = df[df["Tipo"].isin(["Parking", "Local", "Industrial", "Piso"])]
filtroc
```

**INCISO D TODOS LOS REGISTROS CON "PROVINCIA" IGUAL A "GIRONA" Y CON "SUPERFICIE" MAYOR A "100"**

```python
#Filtro Y
filtroD=df[(df["Provincia"] == "Girona") & (df["Superficie"] > 100) ]
filtroD
```

**INCISO E TODOS LOS REGISTROS CON "Precio Venta: mayor a 1,000,000 ó "vendedor" igual a Carmen ó Joaquín**

```python
#Utilizano un filtro or para las 3 condicionales, primero con la columna del Precio y su respectiva cantidad y después para el vendedor con Carmen ó Joaquín
filtroe = df[(df["Precio Venta"] > 1000000) | (df["Vendedor"] == "Carmen") | (df["Vendedor"] == "Joaquín") ]
filtroe
```

**INCISO F TODOS LOS REGISTROS QUE NO SEAN DE "PROVINCIA : BARCELONA"**

```python
#Aplicando el filtro not que muestra todos los valores de la columna Provincia excepto los de "Barcelona"
filtrof=df[~(df["Provincia"] == "Barcelona")]
filtrof
```

**INCISO G TODOS LOS REGISTROS QUE NO SEAN DE "Operación: alquiler" ni de "Tipo: oficina"**

```python
#Aplicando filtro not y un "&" para que se cumplan las dos condicionales de ambas columnas
filtrog = df[~(df["Operación"] == "Alquiler") & ~(df["Tipo"] == "Oficina")]
filtrog
```

```python
#Convertir archivo filtrado a CSV
filtrog.to_csv("filtro_g.csv")
```
